```python
# import pipeline
from transformers import pipeline

# choose the language
lang = "ru"

# create translator object 
translator = pipeline("translation", model=f"Helsinki-NLP/opus-mt-{lang}-en")

# test the model
translator("Российские банки за февраль-март снизили объем закупки золото и других драгоценных металлов.")
```

```python
# make the API call using requests library
import requests

url = "https://free-news.p.rapidapi.com/v1/search"

# define query parameters
query = "золото"

querystring = {"q": query,"lang": lang}

headers = {
	"X-RapidAPI-Host": "free-news.p.rapidapi.com",
	"X-RapidAPI-Key": "SECRET_KEY"
}

response = requests.request("GET", url, headers=headers, params=querystring)

print(response.text)
```

```python
import pandas as pd 
import json
from pandas import json_normalize

# Use json_normalize() to convert JSON to DataFrame
dict = json.loads(response.text)
df = json_normalize(dict['articles']) 
display(df)
```

```python
# extract the summary column and loop through it creating the translations - save them as a list

summary_df = df["summary"]

i = 0
translated = []

for text in summary_df:
  print(20*"-")
  print(i)
  tr = translator(text)
  print(tr)
  translated.append(tr)
  i += 1
  
```

```python
# extract the title column and loop through it creating the translations - save them as a list
title_df = df["title"]

i = 0
translated_title = []

for text in title_df:
  print(20*"-")
  print(i)
  tr = translator(text)
  print(tr)
  translated_title.append(tr)
  i += 1
```

```python
# extract the translated string from the translated object and save to a new list (both title and summary) 

transleted_summary_list = []

for tr in translated:
  #print(20*"-")
  #print(tr[0]["translation_text"])
  transleted_summary_list.append(tr[0]["translation_text"])

transleted_title_list = []

for tr in translated_title:
  #print(20*"-")
  #print(tr[0]["translation_text"])
  transleted_title_list.append(tr[0]["translation_text"])  
```

```python
# reasign the original colums to new ones and asign the translated strings to original colums

df["summary_org"] = df["summary"]
df["summary"] = transleted_summary_list
df["title_org"] = df["title"]
df["title"] = transleted_title_list
df
```

```python
# save the dataframe to a CSV file

df.to_csv(f"{lang}-translated.csv") 
```
