```python
from transformers import pipeline

# create a classifier instance
classifier = pipeline("sentiment-analysis")
# test the classifier
classifier("Elon Musk did some stuff and made some money, but he also created the life of a puppy.")
```

```python
import pandas as pd 

# choose already translated csv file by specifieng the lang paramenter
lang = "us"

#df = pd.read_csv(f"/content/drive/MyDrive/2022.05.24_Final-Project_Shared-Folder/translated/{lang}-translated.csv")
df = pd.read_csv(f"/content/drive/MyDrive/news_{lang}.csv")
df
```

```python
#df.summary()
```

```python
# loop through summary and apply the classifier on each one

summary_df = df["summary"]
i = 0
classified = []

for text in summary_df:
  print(20*"-")
  print(i)
  cl = classifier(text)
  print(cl)
  classified.append(cl)
  i += 1
  
```

```python
# loop through titles and apply the classifier on each one

title_df = df["title"]
i = 0
classified_title = []

for text in title_df:
  print(20*"-")
  print(i)
  cl = classifier(text)
  print(cl)
  classified_title.append(cl)
  i += 1
  
```

```python
#classified
```

```python
labels_list = []
labels_title_list = []

for cl in classified:
  #print(20*"-")
  #print(cl[0]["label"])
  labels_list.append(cl[0]["label"])

for cl in classified_title:
  #print(20*"-")
  #print(cl[0]["label"])
  labels_title_list.append(cl[0]["label"])  
```

```python
df["classification"] = labels_list
df["classification_title"] = labels_title_list
```

```python
df3 = df[["title", "country", "summary", "classification", "classification_title", "language", "topic", "link", "published_date"]]
df3
```

```python
# if the classification of title and summary don't have the same result - mark for manual review 

import numpy as np

df3["review"] = np.where((df3['classification'] != df3['classification_title']), True, False)

df3
```
